# What is a Web API?

1. Application Programming Interfaces define how an application communicates with other programs

2. Web APIs are the most common way to get data from a web application without knowing database details

3. Users can typically pull data from Web APIs with HTTP requests and get data in JSON, CSV, or other formats

4. For example, you send an HTTP request to TMDB that you want the most popular movies in 2019.

5. TMDB will send the 2019 data back to you in JSON format. 

6. You have an API key, then a query string

# Importing and Saving the Movies Dataset

```python
#Here we are creating a list for various movie ids we may pull from the TMDB database

movie_id = [0, 299534, 19995, 140607, 299536, 597, 135397, 420818, 24428, 168259, 99861, 284054, 12445,
           181808, 330457, 351286, 109445, 321612, 260513]
```

```python
#Here our URL has two place holders {}. One for the movie ID and one for the API key 

basic_url = 'https://api.themoviedb.org/3/movie/{}?{}'
```

```python
#Here we are initiating an empty list with a for-loop underneath it that will populate the empty list

json_list = []
#movie_id was initiated above
#Each "movie" will correspond to each movie id in the list above
for movie in movie_id:
    url = basic_url.format(movie, api_key)
    r = requests.get(url)
    #We are setting the response code to 200 since that's desired (meaning the requests are working)
    #So, if the response is not 200, then we will iterate to the next movie id
    if r.status_code != 200:
        continue
    else:
        data = r.json()
        #Each successful movie request pull will be appended to the empty json list
        json_list.append(data)
        #Here we are converting the list into a dataframe
        df = pd.DataFrame(json_list)
```

```python
#Here is a brief look at the dataframe we created after running through the for loop

df.head()
```

```python
#Here we are shrinking our dataframe with just a few selected columns to focus on
#We are sorting them by highest revenue to least revenue

df = df.loc[:, ['title', 'id', 'revenue', 'genres', 'belongs_to_collection', 
                'runtime']].sort_values(by = 'revenue', ascending = False)
```

```python
#Now we are going to export our dataframe to a JSON file for storage
#Orienting by records seems to work well with pandas

df.to_json('movies.json', orient = 'records')
```

```python
#Here we are double checking that we exported the data to JSON correctly by now OPENING the file through JSON

with open('movies.json') as f:
    data = json.load(f)
```

```python
pd.json_normalize(data)
```

```python
pd.json_normalize(data, 'genres', 'title')
```
